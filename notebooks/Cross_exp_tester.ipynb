```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eng_class_df = pd.DataFrame({'language': ['English', 'Spanish', 'Chinese', 'Yorùbá'],
                   'Euphemistic': [853, 1010, 1165, 643],
                   'Non-Euphemistic': [343, 81, 61, 26]})

spa_class_df = pd.DataFrame({'language': ['English', 'Spanish', 'Chinese', 'Yorùbá'],
                   'Euphemistic': [730, 798, 809, 209],
                   'Non-Euphemistic': [466, 293, 416, 460]})

chi_class_df = pd.DataFrame({'language': ['English', 'Spanish', 'Chinese', 'Yorùbá'],
                   'Euphemistic': [867, 1063, 1038, 639],
                   'Non-Euphemistic': [329, 28, 188, 30]})

yor_class_df = pd.DataFrame({'language': ['English', 'Spanish', 'Chinese', 'Yorùbá'],
                   'Euphemistic': [678, 879, 1129, 526],
                   'Non-Euphemistic': [518, 212, 97, 143]})

multi_class_labels = pd.DataFrame({'language': ['English', 'Spanish', 'Chinese', 'Yorùbá'],
                   'Euphemistic': [687, 714, 794, 486],
                   'Non-Euphemistic': [509, 377, 432, 183]})
```

```python
ax = eng_class_df.plot()
spa_class_df.plot(ax=ax)
chi_class_df.plot(ax=ax)
yor_class_df.plot(ax=ax)
```

```python
import itertools

colors = itertools.cycle(["r", "b", "g", "y"])

ax = eng_class_df.plot.scatter(x='Pos', y='Neg',c=next(colors)) #red
spa_class_df.plot.scatter(x='Pos', y='Neg',c=next(colors), ax=ax) # blue
chi_class_df.plot.scatter(x='Pos', y='Neg', c=next(colors), ax=ax) #green
yor_class_df.plot.scatter(x='Pos', y='Neg', c=next(colors), ax=ax) #yellow

plt.title("Cross-Lingual Experiment")
```

```python
# Concatenate the data frames
#data = pd.concat([eng_class_df, spa_class_df, chi_class_df, yor_class_df], axis=1)
concatenated = pd.concat([eng_class_df.assign(dataset='English Classifier'), spa_class_df.assign(dataset='Spanish Classifier'), chi_class_df.assign(dataset='Chinese Classifier'), yor_class_df.assign(dataset='Yorùbá Classifier'), multi_class_labels.assign(dataset='Multilingual Classifier')])


# Create scatter plots
#sns.relplot(x="Pos", y="Neg", hue="Lang", data=concatenated, row=4, col=4)

sns.set_theme()
sns.set_style("darkgrid")

sns.scatterplot(x='Euphemistic', y='Non-Euphemistic', data=concatenated,
                hue='language', style='dataset', s=200).set_title("Cross-Lingual Experiment on Test Data")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
